### CREDENTIALS

```python
IG_LINK = 'https://www.instagram.com/'

USERNAME = input('Enter your Ig Handle: ')
PASSWORD = input('Enter your Ig Password: ')
```

### Module Imports

```python
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from selenium.webdriver.support.select import Select
from selenium.common import exceptions as SCE

import time
```

### Starting session

```python
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

driver.maximize_window()
```

**wait object**

```python
wait = WebDriverWait(driver, 10)
```

# 1) Login to your Instagram Handle

### login function

```python
def log_in(driver, wait, username, password):
    # visit instagram
    driver.get(IG_LINK)
    time.sleep(3)
    
    # wait until log-in form is loaded -> by checking username box
    wait.until(EC.presence_of_element_located((By.NAME, 'username')))
    
    # fill log-in form
    username_box = driver.find_element(By.NAME, 'username')
    username_box.send_keys(username)
    
    password_box = driver.find_element(By.NAME, 'password')
    password_box.send_keys(password)
    
    password_box.submit()
    time.sleep(3)
    
    
    # avoid save login info
    try:
        # wait until or if save login info option appears
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'y3zKF')))
        ele = driver.find_element(By.CLASS_NAME, 'y3zKF')
        ele.click()
        
    except:
        print('No login info save option appears')
    
    
    time.sleep(3)
    

    # avoid turning on notifications
    try:
        # wait until or if turn-on notifications option appears
        wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'HoLwm')))
        ele = driver.find_element(By.CLASS_NAME, 'HoLwm')
        ele.click()
        
    except:
        print('No turn-on notifications option appears')
```

## LOGIN

```python
log_in(driver, wait, USERNAME, PASSWORD)
```

### function to search

```python
def search(driver, wait, text):
    # wait for search_box
    wait.until(EC.presence_of_element_located((By.XPATH, '//input[@placeholder="Search"]')))
    
    # load search box and enter text 
    search_box = driver.find_element(By.XPATH, '//input[@placeholder="Search"]')
    search_box.clear()
    search_box.send_keys(text)
    
    time.sleep(3)
```

## Search food

```python
search(driver, wait, 'food')
```

# 2.) print all the names of the Instagram Handles that are displayed in list after typing “food”

### function to print search result handles

```python
def print_search_result_handles(driver):
    # load search result tags
    ele_list = driver.find_elements(By.XPATH, '//div[@class = "_01UL2"]/div/div/a/div/div[2]/div/div/div/div')

    # iterate ele_list and print search handles
    for ele in ele_list:
        s = ele.get_attribute('innerHTML')
        
        # avoiding blue tick
        if s[0] != '<':
            print(s)
```

## Print handles related to food

```python
print_search_result_handles(driver)
```

# 3.) Open profile of “So Delhi”

### function to open a profile

```python
def open_profile(driver, wait, handle):
    # search text
    search(driver, wait, handle)
    
    # wait for search results to be loaded
    wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class = "_01UL2"]/div/div/a')))
    
    # click on first profile
    driver.find_element(By.XPATH, '//div[@class = "_01UL2"]/div/div/a').click()
    time.sleep(2)
```

## open "So Delhi" handle

```python
open_profile(driver, wait, 'So Delhi')
```

# 4.) a. Follow So Delhi

### function to follow a handle

```python
def follow(driver, wait, handle):
    # open profile
    open_profile(driver, wait, handle)
    
    # wait for search results to be loaded
    wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="react-root"]/section/main/div/header/section/div[1]/div[1]/div/div/div/span/span[1]/button')))
    ele = driver.find_element(By.XPATH, '//*[@id="react-root"]/section/main/div/header/section/div[1]/div[1]/div/div/div/span/span[1]/button')
    
    if ele.text == 'Follow':
        ele.click()
        time.sleep(2)
        print(f"User has Started Following '{handle}'")
    else:
        print(f"User Already Following '{handle}'")
```

## follow "So Delhi"

```python
follow(driver, wait, 'So Delhi')
```

# 4.) b. Unfollow So Delhi

### function to unfollow a handle

```python
def unfollow(driver, wait, handle):
    # open profile
    open_profile(driver, wait, handle)
    
    # wait for search results to be loaded
    wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="react-root"]/section/main/div/header/section/div[1]/div[1]/div/div/div/span/span[1]/button')))
    ele = driver.find_element(By.XPATH, '//*[@id="react-root"]/section/main/div/header/section/div[1]/div[1]/div/div/div/span/span[1]/button')
    
    if ele.text == 'Follow':
        print("Already Unfollowing 'So Delhi'")
    else:
        ele.click()
        time.sleep(1)
        driver.find_element(By.CLASS_NAME, 'aOOlW').click()
        time.sleep(2)
        print("Unfollowing 'So Delhi'")
    
```

## unfollow "So Delhi"

```python
unfollow(driver, wait, 'So Delhi')
```

# 5.) a. Like the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.

### Open profile of "dilsefoodie" :

```python
search(driver, wait, 'dilsefoodie')

# wait for search results to be loaded
wait.until(EC.presence_of_element_located((By.XPATH, '//div[@class = "_01UL2"]/div/div/a')))
driver.find_element(By.XPATH, '//div[@class = "_01UL2"]/div/div/a').click()

time.sleep(2)
```

### fetch first 30 posts element (scroll if required)

```python
# wait until posts are loaded
wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="react-root"]/section/main/div/div[3]/article/div[1]/div/div/div/a')))

# handle page consist 24 posts, so scrolling is mandatory to extract more posts

# current height of page
curr_page_height = driver.execute_script('return document.body.scrollHeight;')

while True:
    # try to fetch 30 posts
    ele_list = driver.find_elements(By.XPATH, '//*[@id="react-root"]/section/main/div/div[3]/article/div[1]/div/div/div/a')[:30]
    
    # break if 30 posts collected
    if len(ele_list) == 30:
        break
        
    # scroll
    driver.execute_script(f'window.scrollTo(0, {curr_page_height});')
    time.sleep(2)
    
    # new page height
    new_page_height = driver.execute_script('return document.body.scrollHeight;')
    
    # if current and new heights are same -> break
    if new_page_height == curr_page_height:
        break
    
    curr_page_height = new_page_height
```

### extract posts link

```python
links = []

for ele in ele_list:
    links.append(ele.get_attribute('href').strip())
    
print('Total posts collected are: ', len(links))
```

### function to like a post

```python
def like_post(driver, link):
    # open link
    driver.get(link)

    # wait until posts are loaded
    wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="react-root"]/section/main/div/div[1]/article/div/div[2]/div/div[2]/section[1]/span[1]/button/div')))

    # load like btn divs
    btn_divs = driver.find_elements(By.XPATH, '//*[@id="react-root"]/section/main/div/div[1]/article/div/div[2]/div/div[2]/section[1]/span[1]/button/div')
    
    if (len(btn_divs)) == 2:
        btn_divs[-1].click()
        return True
        
    return False
        
```

## like the posts

```python
# Like first 30 Posts
for i in range(len(links)):
    if like_post(driver, links[i]):
        print('Post-', i+1, ' : Liked', sep='')
        
    else:
        print('Post-', i+1, ' : Already Liked', sep='')
```

# 5.) b. Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.

### function to unlike a post

```python
def unlike_post(driver, link):
    driver.get(link)
    
    # wait until posts are loaded
    wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="react-root"]/section/main/div/div[1]/article/div/div[2]/div/div[2]/section[1]/span[1]/button/div')))

    # load like btn divs
    btn_divs = driver.find_elements(By.XPATH, '//*[@id="react-root"]/section/main/div/div[1]/article/div/div[2]/div/div[2]/section[1]/span[1]/button/div')
    
    if (len(btn_divs)) == 1:
        btn_divs[-1].click()
        return True
        
    return False
        
```

## unlike the posts

```python
# Unlike first 30 Posts
for i in range(len(links)):
    if unlike_post(driver, links[i]):
        print('Post-', i+1, ' : Unliked', sep='')
    else:
        print('Post-', i+1, ' : Already Unliked', sep='')
```

# 6.) a. Extract the usernames of the first 500 followers of ‘foodtalkindia’ and ‘sodelhi’.

# 500 followers of ‘foodtalkindia’

### function to load n followers

```python
def fetch_n_followers_li_tags(driver, wait, handle, n):
    # open handle profile
    open_profile(driver, wait, handle)
    
    # click on followers btn
    wait.until(EC.presence_of_element_located((By.XPATH, f'//a[@href="/{handle}/followers/"]')))
    driver.find_element(By.XPATH, f'//a[@href="/{handle}/followers/"]').click()
    time.sleep(1)
    
    # loop to check the loading state of 'n' followers, and scroll if required
    while True:
        try:
            # check for 'n'th follower
            driver.find_element(By.XPATH, f'/html/body/div[6]/div/div/div[2]/ul/div/li[{n+20}]')

        except:
            # scroll the followers list, to fetch more followers
            wait.until(EC.presence_of_element_located((By.XPATH, '//ul[contains(@class, "jSC57")]')))
            driver.execute_script('arguments[0].scrollIntoView(0, 100);', driver.find_element(By.XPATH, '//ul[contains(@class, "jSC57")]'))

        else:
            break
          
    # load li tags of 'n' followers
    followers_li_tag_list = driver.find_elements(By.XPATH, '/html/body/div[6]/div/div/div[2]/ul/div/li')[:n]
    
    # return followers_list
    return followers_li_tag_list
```

### fetching 500 followers of foodtalkindia

```python
followers_list = fetch_n_followers_li_tags(driver, wait, 'foodtalkindia', 500)
```

## print handle of 500 followers of foodtalkindia

**along printing handle, I am also storing link of profile of those handles, which user is following**

**this link are later required**

```python
# empty list which store link of all followers which user follows
user_follows_followers = []

for i in range(len(followers_list)):
    # fetch follower's data
    data = followers_list[i].text.split('\n')
    
    # separate the data
    if len(data) == 4:
        handle, verfied_user_status, name, following_status = data
    elif len(data) == 3:
        handle, name, following_status = data
    elif len(data) == 2:
        handle, following_status = data
    else:
        continue
    
    print(f'Follower -{i+1}: {handle}')
    
    # if user follow the follower add its link in 'user_follows_followers' list
    if following_status == 'Following':
        a_tag = driver.find_element(By.XPATH, f'/html/body/div[6]/div/div/div[2]/ul/div/li[{i+1}]/div/div[1]/div[2]/div[1]/span/a')
        user_follows_followers.append([handle, a_tag.get_attribute('href')])
```

# 6.) b. Now print all the followers of “foodtalkindia” that you are following but those who don’t follow you.

### function to whether whether user follows a handle or not (by link)

```python
def is_follow_user(driver, wait, handle_name, link):
    # open link
    driver.get(link)
    time.sleep(2)
    
    # click on following btn
    wait.until(EC.presence_of_element_located((By.XPATH, f'//a[@href="/{handle_name}/following/"]')))
    driver.find_element(By.XPATH, f'//a[@href="/{handle_name}/following/"]').click()
    time.sleep(1)
    
    # extract first following name
    wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[6]/div/div/div[3]/ul/div/li[1]/div/div[2]/div[1]/div/div/span/a')))
    first_following_name = driver.find_element(By.XPATH, '/html/body/div[6]/div/div/div[3]/ul/div/li[1]/div/div[2]/div[1]/div/div/span/a').text
    
    # exit the following ul list
    driver.find_element(By.XPATH, '/html/body/div[6]/div/div/div[1]/div/div[2]/button').click()
    
    
    # if first_following_name is username return True else False
    if first_following_name == USERNAME:
        return True
    else:
        return False
```

## printing followers of foodtalkindia, which user follows, but they don't follow user

```python
for follower in user_follows_followers:
    if not is_follow_user(driver, wait, follower[0], follower[1]):
        print(follower[0])
```

### function to see a user story

```python
# function to see unseen story of handle
def is_present_handle_story(driver, wait, handle):
    try:
        driver.get(IG_LINK)
        
        # wait for story section to be loaded
        wait.until(EC.presence_of_element_located((By.XPATH, f'//button[contains(@aria-label, "{handle}")]')))
        
        story_btn = driver.find_element(By.XPATH, f'//button[contains(@aria-label, "{handle}")]')
        print(f"'{handle}' has share a story today")

        if 'not seen' in story_btn.get_attribute('aria-label'):
            print(f"Seeing {handle}'s story now")
        else:
            print(f"I have already seen the {handle}'s story")
           
        
        # avoid turning on notifications (if appears)
        try:
            # wait until or if turn-on notifications option appears
            wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'HoLwm')))
            ele = driver.find_element(By.CLASS_NAME, 'HoLwm').click()

        except:
            pass
        
        return True
    
    except:
        print(f"'{handle}' has not share any story today")
        return False
```

```python
def view_handle_story(driver, wait, handle):
    open_profile(driver, wait, handle)
    
    # wait for profile image to be loaded and then click it
    wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="react-root"]/section/main/div/header/div/div/span/img')))
    driver.find_element(By.XPATH, '//*[@id="react-root"]/section/main/div/header/div/div/span/img').click()
```

```python
if is_present_handle_story(driver, wait, 'coding.ninjas'):
    print(f"Viewing {handle} story")
    view_handle_story(driver, wait, handle)
```

### function to stop viewing story

```python
def stop_story_view(driver):
    try:
        driver.find_element(By.XPATH, '//*[@id="react-root"]/section/div[3]/button').click()
        print('Running story has been stopped')
    except:
        print('No Story is running')
```

```python
stop_story_view(driver)
```
